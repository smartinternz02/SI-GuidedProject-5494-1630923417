```python
import pandas
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
df.head()
```

# Scatter Graph

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.scatter(df.sepal_length,df.petal_length)
sns.set_style('darkgrid')
```

```python
plt.scatter(df.sepal_length,df.petal_length)
sns.set_style('whitegrid')
plt.show()
```

```python
from matplotlib import pyplot
import seaborn as sns
plt.scatter(df.sepal_length,df.petal_length)
sns.set_style('darkgrid',{'grid.color':'.1'})
plt.show()
```

# Relplot Graph

```python
sns.relplot(data=df,x="sepal_length",y='petal_length')
```

```python
sns.relplot(data=df,x="sepal_length",y='petal_length',hue='species')
```

# KDE PLOT(Kernal Density Estimate)

```python
df.species.unique()
```

```python
sns.kdeplot(df.loc[(df['species']=='setosa'),'sepal_length'],color='b',Label='Setosa')
```

```python
import matplotlib.pyplot as plt
sns.set_style('dark')
sns.kdeplot(df.loc[(df['species']=='setosa'),'sepal_length'],color='b',Label='Setosa')
sns.kdeplot(df.loc[(df['species']=='virginica'),'sepal_length'],shade=True,color='g',Label='Verginica')
```

```python
import matplotlib.pyplot as plt
sns.set_style('dark')
sns.kdeplot(df.loc[(df['species']=='setosa'),'sepal_length'],shade=True,color='b',Label='Setosa')
sns.kdeplot(df.loc[(df['species']=='virginica'),'sepal_length'],shade=True,color='b',Label='Verginica')
```

# boxplot graph

```python
sns.boxplot(x='sepal_length',y='species',data=df)
```

# Bar graph

```python
x=df.species
y=df.sepal_length
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y,color='g',width=0.5,bottom=10,align='center',edgecolor='r',linewidth=2) # customizing
plt.show()
```

# Figure object graph

```python
x=df.species
y=df.sepal_length
plt.plot(x,y)
plt.title("iris")
plt.xlabel('sepal length')
plt.ylabel('species')
plt.grid(True)
plt.plot(x,y,c='r',linestyle='--',linewidth=2,label='species')
plt.legend()
```

```python
fig= plt.figure(figsize=(4,3),facecolor='r',edgecolor='y',linewidth=5)
x=df.species
y=df.sepal_length
plt.plot(x,y,'g')
```

```python
x=df.species
y=df.petal_length
fig,ax=plt.subplots()
ax.plot(y)
```
